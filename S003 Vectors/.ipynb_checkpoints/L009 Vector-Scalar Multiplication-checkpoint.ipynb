```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
v1 = np.array([3, -1])
l = -0.3
v1m = v1 * l

# The plots...
plt.plot([0, v1[0]], [0, v1[1]], 'b', label = 'v_1')
plt.plot([0, v1m[0]], [0, v1m[1]], 'r:', label = 'lamda * v1')

plt.legend()
plt.axis('square')
axlim = max([abs(max(v1)), abs(max(v1m))]) * 1.5 # dynamic axis limits
plt.axis((-axlim, axlim, -axlim, axlim))
plt.grid()
plt.show()
```
