```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# two vectors in R2
v1 = np.array([3, -1])
v2 = np.array([2, 4])

# We need the vectors to be arrays or the + below will be taken as concatenation
v3 = v1 + v2

# The plots...
plt.plot([0, v1[0]], [0, v1[1]], 'b', label = 'v1')
plt.plot([0, v2[0]] + v1[0], [0, v2[1]] + v1[1], 'r', label = 'v2')
plt.plot([0, v3[0]], [0, v3[1]], 'k', label = 'v1 + v2')



plt.legend()
plt.axis('square')
plt.axis((-6, 6, -6, 6))
plt.grid()
plt.show()
```

```python
# two vectors in R3
v1 = np.array([3, -1, 2])
v2 = np.array([2, 4, -1])

# We need the vectors to be arrays or the + below will be taken as concatenation
v3 = v1 + v2

# Plot the 3D vector
fig = plt.figure(figsize = plt.figaspect(1))
ax = fig.gca(projection = '3d')

ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], 'b', linewidth = 3, label = 'v1')
ax.plot([0, v2[0]] + v1[0], [0, v2[1]] + v1[1], [0, v2[2]] + v1[2], 'r', linewidth = 3, label = 'v2')
ax.plot([0, v3[0]], [0, v3[1]], [0, v3[2]], 'k', linewidth = 3, label = 'v1 + v2')

ax.plot([0, 0], [0, 0], [-4, 4], 'k--')
ax.plot([0, 0], [-4, 4], [0, 0], 'k--')
ax.plot([-4, 4], [0, 0], [0, 0], 'k--')
plt.show()

```
